```python
# Importng the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# Importing the dataset
dataset = pd.read_csv('/home/bravin/Documents/Loan Default Prediction Model/vehicle_loans_cleaned.csv', index_col=0)

dataset.head()
```

```python
# inspect columns

dataset.info()
```

```python
category_cols = ['MANUFACTURER_ID', 'STATE_ID', 'DISBURSAL_MONTH', 'PERFORM_CNS_SCORE_DESCRIPTION', 'EMPLOYMENT_TYPE']
dataset[category_cols].dtypes
```

```python
# convert to category

dataset[category_cols] = dataset[category_cols].astype('category')
dataset[category_cols].dtypes

```

```python
small_cols = ['STATE_ID', 'LTV', 'PERFORM_CNS_SCORE', 'DISBURSAL_MONTH', 'LOAN_DEFAULT']

loan_df_sml = dataset[small_cols]
```

```python
# # change LOAN_DEFAULT to 0 and 1
# loan_df_sml['LOAN_DEFAULT'] = loan_df_sml['LOAN_DEFAULT'].map({0:0, 1:1})

loan_df_sml.head()
```

```python
# create x and y variables for model

X = loan_df_sml.drop('LOAN_DEFAULT', axis=1)
y = loan_df_sml['LOAN_DEFAULT']
```

```python
# split data into train and test sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
# initiate model

logmodel = LogisticRegression()

# fit model

logmodel.fit(X_train, y_train)
```
